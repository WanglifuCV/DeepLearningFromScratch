# Study Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.01)
y1 = np.sin(x)
y2 = np.sin(2*x)
```

```python
plt.plot(x, y1, label = 'sin(x)', color = 'red')
plt.plot(x, y2, label = 'sin(2x)', color = 'red',linestyle = '--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.arange(-2, 2, 0.01)
z1 = np.exp(x)
z2 = np.exp(2*x)
z3 = np.exp(0.5*x)
```

```python
plt.plot(x, z1, label='exp^x', color='red', linestyle='-')
plt.plot(x, z2, label='exp^2x', color='blue', linestyle='-')
plt.plot(x, z3, label='exp^0.5x', color='blue', linestyle='--')
plt.legend()
plt.show()
```
